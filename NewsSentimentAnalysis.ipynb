```python

from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
import matplotlib.pyplot as plt

finviz_url = 'https://finviz.com/quote.ashx?t='
tickers = ['AMZN', 'GOOG', 'FB']
```

```python
stock_df = pd.read_csv("stocks.csv")
ticker_list = list(stock_df["Symbol"])
ticker_list
```

```python
news_tables = {}
for i in range(2673):
    try:
        url = finviz_url + ticker_list[i+4000]

        req = Request(url=url, headers={'user-agent': 'my-app'})
        response = urlopen(req)

        html = BeautifulSoup(response, features='html.parser')
        news_table = html.find(id='news-table')
        news_tables[ticker_list[i+4000]] = news_table
        
        print(f"{ticker_list[i+4000]} processed {i}")
        
    except:
        print(f"{ticker_list[i+4000]} didn't process")

parsed_data = []
```

```python
for ticker, news_table in news_tables.items():

    try:
        for row in news_table.findAll('tr'):

            title = row.a.text
            date_data = row.td.text.split(' ')

            if len(date_data) == 1:
                time = date_data[0]
                parsed_data.append([ticker, date, time, title])
            else:
                date = date_data[0]
                time = date_data[1]
                parsed_data.append([ticker, date, time, title])
    except:
        print(f"{ticker} failed")

    
```

```python
df = pd.DataFrame(parsed_data, columns=['ticker', 'date', 'time', 'title'])
```

```python
vader = SentimentIntensityAnalyzer()
```

```python
f = lambda title: vader.polarity_scores(title)['compound']
df['compound'] = df['title'].apply(f)
df['date'] = pd.to_datetime(df.date).dt.date
```

```python
# plt.figure(figsize=(10,8))
mean_df = df.groupby(['ticker', 'date']).mean().unstack()
mean_df = mean_df.xs('compound', axis="columns")
# mean_df.plot(kind='bar')
# plt.show()
```

```python
sentdf = mean_df.fillna(0)
```

```python
def generate_graph(ticker): 
    
    date_ticks =  list(fin_df.loc[fin_df['ticker']==ticker]['Date'])
#     date_ticks = [str(date.date()) for date in dates]                       
    price = list(fin_df.loc[fin_df['ticker']==ticker]['Close'])
    reddit = list(fin_df.loc[fin_df['ticker']==ticker]['reddit_scores'])    #stores the .loc returns as variables 
    sentiment = list(fin_df.loc[fin_df['ticker']==ticker]['sent_scores'])
    
                                            #uses Plotly to graph the lines 
    fig1 = px.line(x=date_ticks, y=price, template= "plotly_dark", title=f"{ticker} price", labels = {"this", "that"}, width=500, height=300)
    fig2 = px.line(x=date_ticks, y=reddit, template= "plotly_dark", title=f"{ticker} Reddit Score", width=500, height=300)
    fig3 = px.line(x=date_ticks, y=sentiment, template= "plotly_dark", title=f"{ticker} Sent Score", width=500, height=300)
                    #the function returns the graph display calls
    return fig1.show() , fig2.show(), fig3.show()
```
